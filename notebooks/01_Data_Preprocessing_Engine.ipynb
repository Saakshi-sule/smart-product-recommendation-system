```python
import pandas as pd
df=pd.read_csv("Ecommerce.csv" , encoding='ISO-8859-1')
df.head()
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
df.dropna(subset=['CustomerID', 'Description'], inplace=True)
df.reset_index(drop=True, inplace=True)
```

```python
df.isnull().sum()
```

```python
df = df[['CustomerID', 'Description', 'Quantity', 'UnitPrice', 'Country']]
df.head()
```

```python
df['Description'] = df['Description'].str.lower()
df['Description'] = df['Description'].str.strip()
df['Description'] = df['Description'].str.replace(r'[^a-zA-Z0-9\s]', '', regex=True)
df.head()
```

```python
#prod-user matrix
product_customer_matrix = df.pivot_table(
    index='Description',       # Products as rows
    columns='CustomerID',      # Customers as columns
    values='Quantity',         # Values = quantity purchased
    fill_value=0               # Fill missing purchases with 0
)
product_customer_matrix.head()
```

```python
from sklearn.metrics.pairwise import cosine_similarity
import pandas as pd

product_similarity = cosine_similarity(product_customer_matrix)

product_similarity_df = pd.DataFrame(
    product_similarity,
    index=product_customer_matrix.index,
    columns=product_customer_matrix.index
)

product_similarity_df.head()
```

```python
def recommend_products(product_name, top_n=5):
    if product_name not in product_similarity_df.columns:
        return "Product not found in dataset."

    similarity_scores = product_similarity_df[product_name].sort_values(ascending=False)

    recommended = similarity_scores.drop(labels=[product_name]).head(top_n)
    return recommended

recommend_products("10 colour spaceboy pen")
```

```python
recommend_products("12 daisy pegs in wood box")
```
